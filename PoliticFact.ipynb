```python
import numpy as np
import pandas as pd
from PolitiFact import PolitiFact
```

```python
df_cand = pd.read_csv("Data/Candidates/Candidates.csv", sep="\t")
df_pf = pd.read_csv("Data/PoliticFact.csv", sep="\t")
```

```python
df_pf.dropna(subset="Name").groupby(["Date"]).count().sort_values(["CheckURL"], ascending=False)
```

```python
def find_cand(row: pd.Series):
    for name in df_cand["Name"]:
        if name in row["Statement"] or name in row["Tags"] or name in row["Poster"]:
            return name 
    return np.NaN
df_pf["Name"] = df_pf.apply(find_cand, axis=1)
```

```python
df_pf.set_index("CheckURL").to_csv("Data/PoliticFact.csv", sep="\t", index_label="CheckURL")
```
